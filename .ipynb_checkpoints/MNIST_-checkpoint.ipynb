### importing libraries

```python
import tensorflow as tf
from tensorflow.keras import layers,models
import matplotlib.pyplot as plt
```

### Load the data

```python
(X_train,y_train),(X_test,y_test)=tf.keras.datasets.mnist.load_data()
```

### Reshape data to include channel

```python
X_train=X_train.reshape(-1,28,28,1)
X_test=X_test.reshape(-1,28,28,1)
```

### Normailize the data

```python
X_train=X_train/255.0
X_test=X_test/255.0
```

### Build Model Architecture

```python
model=models.Sequential([
    layers.Conv2D(32,(3,3),activation='relu',input_shape=(28,28,1)),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64,(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Flatten(),
    layers.Dense(64,activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(10,activation='softmax')
])
```

### compile the model

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

### Train the model

```python
info=model.fit(X_train,y_train,epochs=15,batch_size=64,validation_data=(X_test,y_test))
```

### Evaluate Performance

```python
train_loss,train_acc=model.evaluate(X_test,y_test)
print(f"Train_loss:{train_loss}\nTrain_accuracy:{train_acc}")
```

### Plot Accuracy Graph

```python
plt.plot(info.history['accuracy'],label="train_acc")
plt.plot(info.history['val_accuracy'],label='val_acc')
plt.title("CNN Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy") 
plt.legend()
plt.show()
```

### Vizualize The Prediction

```python
predictions = model.predict(X_test)

plt.figure(figsize=(10, 10))
for i in range(9):
    plt.subplot(3, 3, i + 1)
    plt.imshow(X_test[i].reshape(28, 28), cmap='gray')
    plt.title(f"Predicted: {tf.argmax(predictions[i])}")
    plt.axis('off')
plt.show()
```

### Save the trained Model

```python
model.save("mnist_cnn_model.keras")
```
